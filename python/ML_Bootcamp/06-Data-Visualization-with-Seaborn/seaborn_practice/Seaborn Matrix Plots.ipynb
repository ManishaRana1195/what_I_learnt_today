```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

We can create heatmaps and cluster maps for matric data

```python
flights.head()
```

To plot heatmap, you need corelated data or data in matrix form.
You can create matrix data either by calling corr on data or creating a pivot table

```python
tips_corr = tips.corr()
sns.heatmap(tips_corr,annot=True)
```

```python
sns.heatmap(tips_corr, annot=True,cmap='coolwarm')
```

```python
pivot = flights.pivot_table(index='year',columns='month',values='passengers')
pivot
```

```python
sns.heatmap(pivot)
```

```python
sns.heatmap(pivot, cmap="coolwarm")
```

```python
pivot = flights.pivot_table(index='month', columns='year', values='passengers')
sns.clustermap(pivot, cmap="coolwarm",standard_scale=1)
```

#GRID PLOTS
1. pairGrid
2. facetGrid

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(data=iris,hue='species')
```

```python
grid = sns.PairGrid(iris)
```

```python
import matplotlib as plt
grid = sns.PairGrid(iris,hue='species')
grid.map_diag(sns.distplot)
grid.map_lower(sns.kdeplot)
grid.map_upper(sns.kdeplot)
```
