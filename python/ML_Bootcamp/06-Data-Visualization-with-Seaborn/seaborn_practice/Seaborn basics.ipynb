```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(5) 
```

Following are plots for showing continuous data distribution/pattern

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'], kde=False, bins=10)
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips,kind='kde')
```

```python
sns.jointplot(x=tips['total_bill'],y =tips['tip'], data=tips, kind='hex')
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips,hue="sex", palette="coolwarm")
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

Following are the plots to show categorical data distribution
1. barplot
2. countplot
3. boxplot
4. violinplot
5. stripplot
6. factorplot

```python
sns.barplot(x=tips['sex'],y=tips['tip'],data=tips)
```

```python
sns.barplot(x=tips['tip'],y=tips['day'],data=tips)
```

```python
sns.barplot(x=tips['smoker'],y=tips['tip'],data=tips)
```

```python
import numpy as np
sns.barplot(x=tips['sex'],y=tips['tip'], data=tips,estimator=np.std)
```

```python
sns.countplot(x=tips['size'], data=tips)
```

```python
sns.countplot(x=tips['day'], data=tips)
```

```python
sns.boxplot(x=tips['day'],y=tips['tip'], data=tips)
```

```python
sns.boxplot(x=tips['day'], y=tips['total_bill'], data=tips, hue=tips['sex'])
```

```python
sns.boxplot(x=tips['day'], y=tips['total_bill'], data=tips, hue=tips['smoker'])
```

```python
sns.violinplot(x=tips['day'], y=tips['tip'], data=tips, hue=tips['sex'], split=True)
```

```python
sns.stripplot(x=tips['day'], y=tips['total_bill'], data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker',dodge=True)
```

```python
sns.stripplot(x=tips['day'], y=tips['total_bill'], data=tips, hue=tips['smoker'],dodge=True)
```

```python
sns.catplot(x='day', y='total_bill', data=tips,kind='bar')
```
