```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,20, 10)
y= x ** 2
```

```python
plt.plot(x,y)
plt.xlabel('Income tax')
plt.ylabel('Profit')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2, 2)
plt.plot(x,y, 'b')
```

```python
plt.subplot(2,1,1)                # nrows, ncolumns and index is the position of graph
plt.plot(x,y,'r')

plt.subplot(2,1, 2)
plt.plot(x,y, 'b')
```

```python
#more Object oriented way
fig = plt.figure()
axes = fig.add_axes([0.1,0.1, 1.0, 1.0])
axes.plot(x,y,'g')
smaller = fig.add_axes([0.2,0.6, 0.5,0.5])
smaller.plot(y,x, 'r')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(4,4))
plt.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x,color='purple', linewidth=3)
```

```python
fig.savefig('my_graph.png')
```
